```python
import pandas as pd
import os

data="FashionMNIST"
CWALDP_params = [(49, 4, 8, "Cwaldp49"),(196, 2, 10, "Cwaldp196")]
label_epsilon=2
eps_list=[1,2,3]
test_env="TTS"
model="model2"
all_results=[]
for test_env in ["TTS","UTS"]:
    for model in ["model1","model2","model3"]:
        for eps in eps_list:
            for P, L, cluster_num, cwaldp_name in CWALDP_params:
                filename_cwaldp = f"../../experiments/{data}/CWALDP/CNN/{model}/RR_waldp_L{L}_PI{P}_C{cluster_num}_eps{eps}_env{test_env}_label_noise_{label_epsilon}_time.csv"

                if os.path.exists(filename_cwaldp):
                    df = pd.read_csv(filename_cwaldp, comment="#")
                    accuracy_mean = df["accuracy"].mean()
                    
                    all_results.append({
                        "epsilon": eps,
                        "accuracy_mean": accuracy_mean,
                        "algorithm": cwaldp_name, # Cwaldp1, Cwaldp2 の名前を使用
                        "env":test_env,
                        "model":model
                    })
                    print(f"Loaded {cwaldp_name} for eps={eps}: Acc={accuracy_mean:.4f}")
                else:
                    print(f"File not found for {cwaldp_name} eps={eps}: {filename_cwaldp}")
           
```

```python
results_df = pd.DataFrame(all_results)
print(results_df)
```

```python
select_algorithm="Cwaldp49"
results_df_img=results_df[results_df["algorithm"]==select_algorithm]
# 'env' と 'model' を組み合わせて新しい系列識別子 (Hue) を作成
results_df_img_copy =results_df_img.copy() 

# その後、新しい列を追加
results_df_img_copy['series'] = results_df_img_copy['env'] + '-' + results_df_img_copy['model']

print(results_df_img_copy)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
select_algorithm="Cwaldp196"
results_df_img=results_df[results_df["algorithm"]==select_algorithm]
# 'env' と 'model' を組み合わせて新しい系列識別子 (Hue) を作成
results_df_img_copy =results_df_img.copy() 

# その後、新しい列を追加
results_df_img_copy['series'] = results_df_img_copy['env'] + '-' + results_df_img_copy['model']

print(results_df_img_copy)

plt.figure(figsize=(10, 6))

# 折れ線グラフの描画
sns.lineplot(
    data=results_df_img_copy,
    x='epsilon',               # 横軸: epsilon (1, 2, 3)
    y='accuracy_mean',         # 縦軸: accuracy_mean
    hue='model',               # 色: model1, model2, model3
    style='env',               # 線種: TTSとUTSを実線/点線で区別
    markers=True,              # 各データ点にマーカーを付ける
    dashes=True,               # style='env'を活かすため、dashes=Trueに変更
    linewidth=2
)

# X軸の調整 (1, 2, 3 の整数値のみ表示)
plt.xticks(results_df_img['epsilon'].unique())

# タイトルとラベル
plt.title(f'{select_algorithm} Accuracy Mean vs. Epsilon by Model and Environment', fontsize=12)
plt.xlabel('Privacy Budget $\epsilon$', fontsize=14)
plt.ylabel('Mean Accuracy', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.ylim(0.2, 0.75)
# 凡例を分離（色と線種の説明を分ける）
plt.legend(title='Model', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
